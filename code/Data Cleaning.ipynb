## Data Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in data
df = pd.read_csv('../heart_2020_cleaned.csv')
df.head()
```

```python
df.nunique()
```

```python
df = df[df['HeartDisease'] != 'w r']
arr = ['HeartDisease','Diabetic','Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'KidneyDisease', 'SkinCancer', 'Asthma', 'PhysicalActivity']             
for i in arr:
    df[i] = df[i].map({'Yes': 1, 'No': 0})
df.head()
```
